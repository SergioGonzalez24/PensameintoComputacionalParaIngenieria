## Problema  2

Genere una función en Python que reciba como argumentos de entrada dos listas con números que pueden estar en desorden, que las una y los ordene en forma descendente en una nueva lista y que regrese ésta nueva lista. Además el programa deberá pedir las listas y mostrar la nueva lista y el número de elementos de ésta.

Ejemplos de ejecución:

```
[3,0,9]
[-2,18,2,6]
[18, 9, 6, 3, 2, 0, -2]
7

[]
[0,-2,-3]
[0, -2, -3]
3

[1,2,3]
[0,2]
[3, 2, 2, 1, 0]
5

```

````python
def ordenadas(listaA,listaB):
    '''Escribir documentación aquí'''
    C=listaA+listaB #[1,2,3]+[0,2]= [1,2,3,0,2]
    C.sort(reverse=True)#descendente
    return C,len(C) #  [3, 2, 2, 1, 0] , 5

def main():
    A=eval(input()) #[1,2,3]
    B=eval(input()) #[0,2]
    (C,n_ele) = ordenadas(A,B) #  [3, 2, 2, 1, 0] , 5
    print(C)
    print(n_ele)

main()
````

## Problema 3
Escribir una función en Python que reciba como parámetros de entrada dos listas, que verifique si son de la misma longitud; si son de la misma longitud que muestre una nueva lista, donde cada elemento es la suma de los correspondientes en las dos listas, si no son de la misma longitud que regrese 0. El programa deberá pedir las listas y mostrar la nueva lista o regresar el cero.

Ejemplos de ejecución:
```
[1,2,3]
[4,5,6]
[5, 7, 9]

[11,0]
[1,2,3]
0

```

````python
def sumav(lista1,lista2):
    '''Aquí va la documentación'''
    if len(lista1)==len(lista2):
        suma=[]
        for k in range(len(lista1)): #recorro listas por índices,k=0,1,2
            suma.append( lista1[k]+lista2[k] )
        return suma
    else:
        return 0
    
def main():
    A=eval(input())
    B=eval(input())
    print( sumav(A,B) )
    
main()
````
